```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

amex_df = pd.read_csv('../chapter4/gen/amex_df.csv',
                           parse_dates=True,
                           index_col=0)
```

```python
amex_df.close.plot()
# ma 30
amex_df['close'].rolling(30).mean().plot()
# ma 60
amex_df['close'].rolling(60).mean().plot()
# ma 90
amex_df['close'].rolling(90).mean().plot()
# loc=best
plt.legend(['close',
            '30 mv',
            '60 mv',
            '90 mv'],
           loc='best')
```
