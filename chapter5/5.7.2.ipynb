```python

# 1. Visual Gold Ratio
# Find the max 
cs_max = amex_df.close.max()
# Find the min
cs_min = amex_df.close.min()

sp382 = (cs_max-cs_min) * 0.382 + cs_min
sp618 = (cs_max-cs_min) * 0.618 + cs_min

print('Visual Gold Ratio 382: {}'.format(round(sp382, 2)))
print('Visual Gold Ratio 618: {}'.format(round(sp618, 2)))

```

```python
"""
Use 0.20, 0.25, 0.30 to replace 0.382
Use 0.70, 0.80, 0.90 to replace 0.618
"""
rate382 = [0.20, 
        0.25, 
        0.30]
rate618 = [0.70,
           0.80,
           0.90]

def find_percent_point(percent, y_org, want_max):
    """
    :param percent: 
    :param y_org: 
    :param want_max: 
    :return: 
    """
    cs_max = y_org.max()
    cs_min = y_org.min()
    
    maxmin_mum = np.maximum if want_max else np.minimum
    
    return maxmin_mum(
        # Statistically Gold
        stats.scoreatpercentile(y_org,
                                np.round(percent * 100,
                                         1)),
        # Visual GOld
        (cs_max - cs_min) * percent + cs_min)

result = []
result.append(
    (0.382, 0.618, round(golden.upper618 - golden.lower382, 2)))

import itertools
for (buy, sell) in itertools.product(rate382, rate618):
    profit_below = find_percent_point(buy, amex_df.close, False)
    profit_above = find_percent_point(sell, amex_df.close, True)
    result.append((buy, sell, round(profit_above - profit_below, 2)))

result = np.array(result)
result
```

```python
"""
Either use scatter(), 
or use mpl_toolkts.mplot3d
"""

# Using catter
cmap = plt.get_cmap('jet', 20)
cmap.set_under('gray')
_, ax = plt.subplots(figsize=(8, 5))

cax = ax.scatter(result[:, 0],
                 result[:, 1],
                 c=result[:, 2],
                 cmap=cmap,
                 vmin=np.min(result[:, 2]),
                 vmax=np.max(result[:, 2]))
plt.colorbar(cax, label='max profit', extend='min')
plt.grid(True)
plt.xlabel('buy rate')
plt.ylabel('sell rate')
plt.show()
```
