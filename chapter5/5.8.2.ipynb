```python

"""
Generally just use TA-Lib
"""

# 5.8.2 Visualizing MACD

import pandas as pd
import numpy as np
from abupy import ABuSymbolPd
import matplotlib.pyplot as plt
import talib

amex_df = pd.read_csv('../chapter4/gen/amex_df.csv',
                           parse_dates=True,
                           index_col=0)
```

```python
import talib

# Calculate atr14
atr14 = talib.ATR(amex_df.high.values,
                  amex_df.low.values,
                  amex_df.close.values,
                  timeperiod=14)

# Calculate atr21
atr21 = talib.ATR(amex_df.high.values,
                  amex_df.low.values,
                  amex_df.close.values,
                  timeperiod=21)
pd.DataFrame({
    'close': amex_df.close,
    'atr14': atr14,
    'atr21': atr21
}).plot(subplots=True,
        grid=True)
```
