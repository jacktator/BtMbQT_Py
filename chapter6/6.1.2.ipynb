```python

"""
There're 3 ways to measure Goodness of Fit
1. MAE: Easiest to understand
2. MSE: Heavy penalty for outliers
3. RMSE: Better Understand (Follow norm)
"""
```

```python

# 6.1.1 Linear Regression

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

amex_df = pd.read_csv('../chapter4/gen/amex_df.csv',
                           parse_dates=True,
                           index_col=0)
```

```python

amex_close = amex_df.close
# x as independent variable
x = np.arange(0, amex_df.shape[0])
# y as dependent variable
y = amex_close.values
```

```python

# 6.1.2 Polynomial

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

amex_df = pd.read_csv('../chapter4/gen/amex_df.csv',
                           parse_dates=True,
                           index_col=0)
```

```python
import itertools
from sklearn import metrics

# Generate 9 subplots
_ ,axs = plt.subplots(nrows=3,
                      ncols=3,
                      figsize=(15, 15))
# Turn 3 x 3 into a list
axs_list = list(itertools.chain.from_iterable(axs))
# 1~9 Polynomial
poly = np.arange(1, 10, 1)
for p_cnt, ax in zip(poly, axs_list):
    # Polynomial Regression by polynomial.Chebyshev.fit()
    p = np.polynomial.Chebyshev.fit(x, y, p_cnt)
    # Use p to get y_fit
    y_fit = p(x)
    # Measure MSE
    mse = metrics.mean_squared_error(y, y_fit)
    # Set titles
    ax.set_title('{} poly MSE={}'.format(p_cnt,
                                         mse))
    ax.plot(x, 
            y, 
            '', 
            x, 
            y_fit, 
            'r.')
```
