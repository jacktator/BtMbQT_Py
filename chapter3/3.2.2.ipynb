```python

import numpy as np

a_investor = np.random.normal(100, 50, (100, 1))
b_investor = np.random.normal(100, 20, (100, 1))
```

```python

print('Investor A Mean ${0:.2f}, STD {1:.2f}, VAR {2:.2f}'.format(a_investor.mean(), a_investor.std(), a_investor.var()))

print('Investor B Mean ${0:.2f}, STD {1:.2f}, VAR {2:.2f}'.format(b_investor.mean(), b_investor.std(), b_investor.var()))
```

```python

a_mean = a_investor.mean()

a_std = a_investor.std()

from matplotlib import pyplot as plt
plt.plot(a_investor)

plt.axhline(a_mean + a_std, color='r')

plt.axhline(a_mean, color='y')

plt.axhline(a_mean - a_std, color='g')
```

```python
 
from scipy.stats import kurtosis
from scipy.stats import skew

print( 'A: excess kurtosis of normal distribution (should be 0): {}'.format( kurtosis(a_investor) ))
print( 'A: skewness of normal distribution (should be 0): {}'.format( skew(a_investor) ))
```
