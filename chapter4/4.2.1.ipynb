```python

from abupy import ABuSymbolPd

amex_df = ABuSymbolPd.make_kl_df('AXP', n_folds=2)

amex_df.tail()

amex_df.to_csv('./gen/amex_df.csv', 
               columns=amex_df.columns,
               index=True)
```

```python

amex_df[['close', 'volume']].plot(subplots=True, style=['r', 'g'])

print("Amex Mean {}".format(amex_df[['p_change']].mean()))
print("Amex STD  {}".format(amex_df[['p_change']].std()))
```
