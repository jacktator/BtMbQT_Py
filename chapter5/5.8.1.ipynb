```python

"""
Generally just use TA-Lib
"""

# 5.8.1 Visualizing MACD

import pandas as pd
import numpy as np
from abupy import ABuSymbolPd
import matplotlib.pyplot as plt
import talib

amex_df = pd.read_csv('../chapter4/gen/amex_df.csv',
                           parse_dates=True,
                           index_col=0)
```

```python

kl_index = amex_df.index

dif, dea, bar = talib.MACD(amex_df.close.values,
                            fastperiod=12,
                            slowperiod=26,
                            signalperiod=9)
plt.plot(kl_index,
         dif,
         label='macd dif')
plt.plot(kl_index, 
         dea, 
         label='signal dea')
bar_red = np.where(bar > 0, bar, 0)
bar_green = np.where(bar < 0, bar, 0)
# Drawing bar > 0
plt.bar(kl_index,
        bar_red,
        facecolor='red',
        label='hist bar')
# Drawing bar < 0
plt.bar(kl_index,
        bar_green,
        facecolor='green',
        label='hist bar')
plt.legend(loc='best')
```
